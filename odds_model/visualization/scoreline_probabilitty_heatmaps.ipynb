```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
all_scorelines = pd.read_pickle('../data/scoreline_probabilities.pkl')
```

```python
for idx, home_team in enumerate(all_scorelines.home_team.unique()):
    away_team =  all_scorelines.away_team.unique()[idx]
    
    data = pd.pivot_table(
        all_scorelines[all_scorelines.home_team == home_team],
        index='home_score',
        columns='away_score',
        values='odds'
    )
    
    data = 1/data #convert to implied probability
    fig, ax = plt.subplots(figsize=(15,15))
    ax = sns.heatmap(data, annot=True, fmt='.2%')
    ax.set_title(f'{home_team} vs {away_team}')
    plt.show()
```
